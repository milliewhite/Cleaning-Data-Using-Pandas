```python
# Read data from CSV and clean it for easier analysis of the data
# data is from https://nces.ed.gov/programs/digest/d20/tables/dt20_104.30.asp
```

```python
import pandas as pd
df = pd.read_csv('nces.csv')
df.head(50)
```

```python

df.columns
```

```python
df = df.drop(df.index[0])
df
```

```python
df.columns
```

```python
#keep columns with numbers in thousands
#get rid of cols with standard deviation
#get rid of columns with all null values
#make list of columns to keep
keep_cols = ['Unnamed: 0','Unnamed: 1','Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13','Unnamed: 15','Unnamed: 17','Unnamed: 19']
necs_df = df.copy()
necs_df = necs_df[keep_cols]
necs_df
```

```python
#drop rows where EVERY row is NaN
necs_df = necs_df.dropna(axis = 0, how = 'all')
necs_df
```

```python
#create dict with old and new column names
new_cols = {'Unnamed: 0':'Age','Unnamed: 1':'Total','Unnamed: 3':'Elementary school','Unnamed: 5':'HS_1-3yrs','Unnamed: 7':'HS_4yrs_no_completion','Unnamed: 9':'HS_completion','Unnamed: 11':'Some_college_No_degree','Unnamed: 13':'Associates','Unnamed: 15':"Bachelor's",'Unnamed: 17':"Master's",'Unnamed: 19':'Doctorate'}
#Rename the columns
necs_df.rename(columns = new_cols,inplace=True)
necs_df
```

```python
#remove rows 1,2 and 3..for some reason I had to put index 0,1 and 2...
necs_df = necs_df.drop(df.index[[0,1,2]])
necs_df
```

```python
necs_df.reset_index()
```

```python
#reset index and drop the old index so that it does not become a new column
necs_df = necs_df.reset_index(drop = True)
necs_df
```

```python
#values in the first column
necs_df['Age'].unique()
```
